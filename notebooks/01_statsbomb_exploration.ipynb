```python
import numpy as np
import pandas as pd
```

```python
def add_spatial_features(shots):

    shots = shots.copy()

    shots["x"] = shots["location"].apply(
        lambda loc: loc[0]
    )

    shots["y"] = shots["location"].apply(
        lambda loc: loc[1]
    )

    shots["distance"] = np.sqrt(
        (120 - shots["x"])**2 +
        (40 - shots["y"])**2
    )

    left_post_y = 44
    right_post_y = 36

    shots["angle"] = np.abs(
        np.arctan2(
            left_post_y - shots["y"],
            120 - shots["x"]
        )
        -
        np.arctan2(
            right_post_y - shots["y"],
            120 - shots["x"]
        )
    )

    shots["angle_deg"] = np.degrees(
        shots["angle"]
    )

    return shots
```

```python
def add_target(shots):

    shots = shots.copy()

    shots["goal"] = (
        shots["shot_outcome"] == "Goal"
    ).astype(int)

    return shots
```

```python
def add_boolean_features(shots):

    shots = shots.copy()

    shots["shot_first_time"] = (
        shots["shot_first_time"]
        .fillna(False)
        .astype(int)
    )

    shots["shot_one_on_one"] = (
        shots["shot_one_on_one"]
        .fillna(False)
        .astype(int)
    )

    return shots
```

```python
def nearest_defender_distance(
    freeze_frame,
    shot_x,
    shot_y
):

    if not isinstance(freeze_frame, list):
        return np.nan

    defenders = []

    for player in freeze_frame:

        if player["teammate"]:
            continue

        if player["position"]["name"] == "Goalkeeper":
            continue

        px = player["location"][0]
        py = player["location"][1]

        d = np.sqrt(
            (px - shot_x)**2 +
            (py - shot_y)**2
        )

        defenders.append(d)

    if len(defenders) == 0:
        return np.nan

    return min(defenders)
```

```python
def add_defender_feature(shots):

    shots = shots.copy()

    shots["nearest_defender_distance"] = shots.apply(
        lambda row:
        nearest_defender_distance(
            row["shot_freeze_frame"],
            row["x"],
            row["y"]
        ),
        axis=1
    )

    return shots
```

```python
def select_xg_features(shots):

    return shots[
        [
            "distance",
            "angle_deg",
            "nearest_defender_distance",
            "shot_body_part",
            "shot_type",
            "shot_first_time",
            "shot_one_on_one",
            "goal"
        ]
    ].copy()
```

```python
def process_match(match_id):

    events = sb.events(match_id=match_id)

    shots = events[
        events["type"] == "Shot"
    ].copy()

    shots = add_spatial_features(shots)

    shots = add_target(shots)

    shots = add_boolean_features(shots)

    shots = add_defender_feature(shots)

    shots = select_xg_features(shots)

    return shots
```
